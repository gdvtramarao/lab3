```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
df=pd.read_csv(r"D:\\extracted_features_charrec_dense.csv")
df
```

```python
means = df.mean()
centroid1 = means.loc['6']
centroid2 = means.loc['12']
display= np.linalg.norm(centroid1 - centroid2)

print(f"Euclidean distance between the mean vectors of class A and B is {display:.2f}")
```

```python
var=df.var()
cent1=var.loc['12']
cent2=var.loc['11']
display=np.linalg.norm(cent1-cent2)
print(f"Euclidean distance between the var vectors of class A,B is {display:.2f}")
```
